## 1Ô∏è‚É£ Tuples

A **tuple** is an ordered, immutable collection of values.

- **Ordered**: elements have positions
- **Immutable**: elements cannot be changed after creation

Examples:
- `(3, 7, 10)`
- `(1, "apple", 3.14)`


```python
# Run this cell

t = (3, 7, 10)
t
```

## 2Ô∏è‚É£ Tuples vs Lists

| Feature | List | Tuple |
|-------|------|-------|
| Mutable | Yes | No |
| Syntax | `[ ]` | `( )` |
| Best for fixed data | No | Yes |

Once created, tuple values cannot be reassigned.


```python
# Lists are mutable
lst = [1, 2, 3]
lst[0] = 99
lst
```

```python
# Tuples are immutable
tup = (1, 2, 3)
tup[0] = 99  # Uncomment to observe the error
tup
```

## Exercise 1 ‚Äî Most Frequent Characters

Write a function `most_frequent(s)` that:
- counts how many times each character appears
- stores results as a **list of (character, count) tuples**
- prints the characters in decreasing order of frequency
- use : sorted(nums, reverse=True)


```python


def most_frequent(s):
    # TODO: use a list of tuples (no dictionaries)
    letters = []
    for letter in s:
      letters.append(letter)
      count = 0
      for letter in letters:
        count += 1
      print(count)

# Test string
most_frequent('banana')
```

## Exercise 2 ‚Äî Word Frequencies

Given a list of words, count how often each word appears.
Return a list of `(word, count)` tuples sorted by count (descending).
- use : sorted(nums, reverse=True)



```python
def word_frequencies(words):
  my_tuple = []
  for word in words:
    my_tuple.append(word)
    for row in my_tuple:
      if word in row:
        my_tuple.append("1")
      else:
        continue
      return my_tuple

words = ['the', 'cat', 'the', 'dog']
print(word_frequencies(words))
```

## Exercise 3 ‚Äî Most Common Item

Write a function that takes a list of `(item, count)` tuples
and returns the tuple with the highest count.


```python
def most_common(counts):
  maxi = None
  for row in counts:
    count = 0
    for col in row:
      count += 1
      if count > 0:
        if maxi == None or col > maxi:
          maxi = row
      else:
        continue
      return maxi

counts = [("Apple", 5), ("Banana", 6), ("Cookie", 4)]
maximum = most_common(counts)
print(maximum)
```
